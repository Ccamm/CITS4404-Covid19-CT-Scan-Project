```python
%matplotlib inline
import glob, os, shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import load_img, img_to_array, array_to_img, ImageDataGenerator
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import shuffle
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, InputLayer
from tensorflow.keras.models import Sequential
from tensorflow.keras import optimizers
import random
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)

SEED = 1234
EPOCHS = 100

DATASET_LOC = os.path.join("..", "COVID-CT")
IMAGES_LOC = os.path.join(DATASET_LOC, "Images-processed")
COVID_IMAGES_LOC = os.path.join(IMAGES_LOC, "CT_COVID")
NOCOVID_IMAGES_LOC = os.path.join(IMAGES_LOC, "CT_NonCOVID")
DATASPLIT_LOC = os.path.join(DATASET_LOC, "Data-split")

# Full size of images but too big
# IMAGE_DIM = (335, 580)
IMAGE_DIM = (128, 128)

random.seed(a=SEED)
np.random.seed(SEED)
```

```python
def get_data_filenames(is_covid):
    """
    Returns a list of the image filenames for the phase of the training.
    
    Parameters:
        image_loc: str
            folder location of the images
        is_covid: bool
            is the file list for covid or non-covid data
            
    Returns:
        A tuple containing the filnames of images for test, train and validation.
    """
    data_type_str = "COVID" if is_covid else "NonCOVID"
    base_dir = os.path.join(DATASPLIT_LOC, data_type_str)
    
    with open(os.path.join(base_dir, "testCT_{data_type}.txt".format(data_type=data_type_str)), 'r') as f:
        test_filenames = f.read().split('\n')[:-1]
        test_filenames = [os.path.join(IMAGES_LOC, "CT_{data_type}".format(data_type=data_type_str), filename) 
                              for filename in test_filenames]
    
    with open(os.path.join(base_dir, "trainCT_{data_type}.txt".format(data_type=data_type_str)), 'r') as f:
        train_filenames = f.read().split('\n')[:-1]
        train_filenames = [os.path.join(IMAGES_LOC, "CT_{data_type}".format(data_type=data_type_str), filename) 
                              for filename in train_filenames]
        
    with open(os.path.join(base_dir, "valCT_{data_type}.txt".format(data_type=data_type_str)), 'r') as f:
        val_filenames = f.read().split('\n')[:-1]
        val_filenames = [os.path.join(IMAGES_LOC, "CT_{data_type}".format(data_type=data_type_str), filename) 
                              for filename in val_filenames]
    
    return test_filenames, train_filenames, val_filenames

covid_test_train_val_filenames = get_data_filenames(True)
noncovid_test_train_val_filenames = get_data_filenames(False)

covid_test_train_val_filenames
```

```python
def create_test_train_val_datasets(covid_test_train_val_filenames, noncovid_test_train_val_filenames):
    """
    
    """
    
    def load_images(test_train_val_filenames, data_type_str):
        """
        Loads the images as numpy arrays that are standardized.

        Parameters:
            test_train_val_filenames: tuple
                the tuple containing the lists for testing, training and validation
        """
        test_images = [img_to_array(load_img(filename, target_size=IMAGE_DIM)).astype('float32')/255
                          for filename in test_train_val_filenames[0]]

        train_images = [img_to_array(load_img(filename, target_size=IMAGE_DIM)).astype('float32')/255
                          for filename in test_train_val_filenames[1]]

        val_images = [img_to_array(load_img(filename, target_size=IMAGE_DIM)).astype('float32')/255
                          for filename in test_train_val_filenames[2]]
        
        test_values = [data_type_str for _ii in range(len(test_images))]
        train_values = [data_type_str for _ii in range(len(train_images))]
        val_values = [data_type_str for _ii in range(len(val_images))]
        
        return (test_images, test_values), (train_images, train_values), (val_images, val_values)
    
    def create_dataset(covid_list, noncovid_list):
        combine_labels = np.array(covid_list[1] + noncovid_list[1])
        combine_data = np.array(covid_list[0] + noncovid_list[0])
        print(combine_labels.shape, combine_data.shape)
        combine_data, combine_labels = shuffle(combine_data, combine_labels)
        return combine_data, combine_labels
    
    covid_test_train_val_images = load_images(covid_test_train_val_filenames, "COVID")
    noncovid_test_train_val_images = load_images(noncovid_test_train_val_filenames, "NonCOVID")
    
    test_data, test_labels = create_dataset(covid_test_train_val_images[0], noncovid_test_train_val_images[0])
    train_data, train_labels = create_dataset(covid_test_train_val_images[1], noncovid_test_train_val_images[1])
    val_data, val_labels = create_dataset(covid_test_train_val_images[2], noncovid_test_train_val_images[2])
    
    return test_data, test_labels, train_data, train_labels, val_data, val_labels

test_data, test_labels, train_data, train_labels, val_data, val_labels = create_test_train_val_datasets(
    covid_test_train_val_filenames,
    noncovid_test_train_val_filenames
)

print("Training Data Dimensions:", train_data.shape)
print("Testing Data Dimensions:", test_data.shape)
print("Validation Data Dimensions", val_data.shape)
print()

plot_dim = (5, 5)
num_of_samples = plot_dim[0]*plot_dim[1]
fig, ax = plt.subplots(plot_dim[0], plot_dim[1], figsize=(10,10))

print("First {} Images From Training Set".format(num_of_samples))
for xx in range(num_of_samples):
    ii = xx // plot_dim[1]
    jj = xx % plot_dim[1]
    
    ax[ii][jj].axis('off')
    ax[ii][jj].set_title(train_labels[xx])
    ax[ii][jj].imshow(array_to_img(train_data[xx]))
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(train_labels)

train_labels_enc = label_encoder.transform(train_labels)
test_labels_enc = label_encoder.transform(test_labels)
val_labels_enc = label_encoder.transform(val_labels)

for ii in range(5):
    print(train_labels_enc[ii], train_labels[ii])
```

```python
input_shape = (IMAGE_DIM[0], IMAGE_DIM[1], 3)

res_net = ResNet50(include_top=False, weights="imagenet", input_shape=input_shape, pooling="max")

# res_net_output = res_net.layers[-1].output
# res_net_output = keras.layers.Flatten()(res_net_output)

# res_net = Model(res_net.input, res_net_output)

for layer in res_net.layers:
    if "conv5_block3" in layer.name:
        layer.trainable = True
    else:
        layer.trainable = False
    
res_net.trainable = True

layers = [(layer, layer.name, layer.trainable) for layer in res_net.layers]
pd.DataFrame(layers, columns=['Layer Type', 'Layer Name', 'Layer Trainable'])
```

```python
res_net.summary()
```

```python
DROP_OUT_RATE = 0.3

cnn_model = Sequential([
    res_net,
    Dense(512, activation="relu"),#, input_shape=input_shape),
    Dropout(DROP_OUT_RATE),
    Dense(512, activation="relu"),
    Dropout(DROP_OUT_RATE),
    Dense(1, activation='sigmoid')
])
# cnn_model.add(res_net)
# cnn_model.add(Dense(512, activation="relu",  
#                             input_shape=input_shape))
# cnn_model.add(Dropout(DROP_OUT_RATE))
# cnn_model.add(Dense(512, activation="relu"))
# cnn_model.add(Dropout(DROP_OUT_RATE))
# cnn_model.add(Dense(1, activation='sigmoid'))

cnn_model.compile(loss='binary_crossentropy',
                  optimizer=optimizers.RMSprop(lr=2e-5),
                  metrics=["accuracy"])

cnn_model.summary()
```

```python
data_gen = ImageDataGenerator(width_shift_range=0.1, height_shift_range=0.1, zoom_range=0.3, rotation_range=50,
                                 shear_range=0.2, horizontal_flip=True, fill_mode='nearest')
data_gen.fit(train_data)
val_data_gen = ImageDataGenerator(rescale=1./255)
val_data_gen.fit(val_data)

train_gen = data_gen.flow(train_data, train_labels_enc, batch_size=64)
val_gen = val_data_gen.flow(val_data, val_labels_enc, batch_size=64)
```
