```python
%matplotlib inline
import glob, os, shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import random, json
from sklearn.metrics import confusion_matrix, f1_score
```

## Loading Dataset and Organising for Training, Testing and Validation

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)

SEED = 1234
INITIAL_EPOCHS = 100
FINETUNE_EPOCHS = 50
IMAGE_DIM = (160, 160)
BATCH_SIZE = 32

DATASET_LOC = os.path.join("dataset")
train_dataset = tf.keras.preprocessing.image_dataset_from_directory(DATASET_LOC,
                                                                        shuffle = True,
                                                                        batch_size=BATCH_SIZE,
                                                                        image_size=IMAGE_DIM,
                                                                        seed=SEED)

random.seed(a=SEED)
np.random.seed(SEED)
```

```python
class_names = train_dataset.class_names

plt.figure(figsize=(10, 10))
for images, labels in train_dataset.take(1):
    for ii in range(9):
        ax = plt.subplot(3, 3, ii + 1)
        plt.imshow(images[ii].numpy().astype("uint8"))
        plt.title(class_names[labels[ii]])
        plt.axis("off")
```

We split the dataset such that 1/3 of the images are placed into a validation dataset and another 1/3 from the validation data set is then put into a testing dataset.

```python
# 1/3 of the images are put in a validation dataset, then 1/3 from the validation dataset into a test set
train_batches = tf.data.experimental.cardinality(train_dataset)
val_dataset = train_dataset.take(train_batches // 3)
train_dataset = train_dataset.skip(train_batches // 3)

val_batches = tf.data.experimental.cardinality(val_dataset)
test_dataset = val_dataset.take(val_batches // 3)
val_dataset = val_dataset.skip(val_batches // 3)

print('Number of train batches: %d' % tf.data.experimental.cardinality(train_dataset))
print('Number of validation batches: %d' % tf.data.experimental.cardinality(val_dataset))
print('Number of test batches: %d' % tf.data.experimental.cardinality(test_dataset))
```

Prefetches images to prevent I/O blocking and slower time preformance.

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
val_dataset = val_dataset.prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.prefetch(buffer_size=AUTOTUNE)
```

## Data Augmentation

Since the dataset is relatively small in comparison to other datasets that CNNs are trained on (eg. ImageNet), the size of the training dataset needs to be augmented.

This is done by applying a random flip and rotation to the images in the dataset to provide more training instances.

```python
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.experimental.preprocessing.RandomFlip('horizontal'),
  tf.keras.layers.experimental.preprocessing.RandomRotation(0.2),
])
```

```python
for image, labels in train_dataset.take(1):
    plt.figure(figsize=(10, 10))
    first_image = image[0]
    for ii in range(9):
        ax = plt.subplot(3, 3, ii + 1)
        aug_image = data_augmentation(tf.expand_dims(first_image, 0))
        plt.imshow(aug_image[0] / 255)
        plt.axis('off')
```

## Creating the CNN Model

Firstly a preprocess input layer is created for the MobileNetV2 CNN that we are using which processes the shape of the input to be the same shape that the MobileNEtV2 CNN was trained on.

Then the MobileNetV2 pre-trained model on the ImageNet dataset is imported and all of the layers are 'frozen' (not being trained).

### **Go to the [Results](#Results) section if you do not want to train the model and just to view the results from training**

```python
# Function for rescaling the images for input into the MobileNetV2 from uint8 [0, 255] to float32 [-1, 1]
preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input
```

```python
IMAGE_SHAPE = IMAGE_DIM + (3,)
base_model = tf.keras.applications.MobileNetV2(input_shape=IMAGE_SHAPE,
                                                   include_top=False,
                                                   weights='imagenet')

base_model.trainable = False
base_model.summary()
```

Using the Keras Functional API, the entire model with the dense layer on top of the MobileNetV2 CNN is created.

```python
def get_model():
    image_batch, label_batch = next(iter(train_dataset))
    feature_batch = base_model(image_batch)

    global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
    feature_batch_average = global_average_layer(feature_batch)

    prediction_layer = tf.keras.layers.Dense(1)
    prediction_batch = prediction_layer(feature_batch_average)

    inputs = tf.keras.Input(shape=IMAGE_SHAPE)
    augmented_inputs = data_augmentation(inputs)
    processed_inputs = preprocess_input(augmented_inputs)
    mobilenet_output = base_model(processed_inputs, training=False)
    global_average_output = global_average_layer(mobilenet_output)
    dropout_output = tf.keras.layers.Dropout(0.2)(global_average_output)
    outputs = prediction_layer(dropout_output)

    return tf.keras.Model(inputs, outputs)

model = get_model()
model.summary()
```

# Training Part 1
### Training the Dense Neural Network on the 'Frozen' MobileNetV2 CNN

The Dense layer of the Neural Network is trained while the weights of the MobileNetV2 network are frozen. This trains the dense layer to be able to interpret the feature map from the MobileNetV2 so when the MobileNetV2 is finetuned the dense layer would only have to make minor adjustments to its weights.

```python
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.0001),
                  loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                  metrics=['accuracy'])
```

```python
history = model.fit(train_dataset,
                        epochs=INITIAL_EPOCHS,
                        validation_data = val_dataset)
```

# Training Part 2
## Un-freezing the Top 54 Layers of the MobileNetV2 and Training the Weights Along with the DNN Component

For another 50 epochs the top 54 layers of the MobileNetV2 is fine tuned to extract appropriate features from the CT scans.

```python
FINE_TUNE_AT = 100

base_model.trainable = True

for layer in base_model.layers[:FINE_TUNE_AT]:
    layer.trainable = False
    
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.0001),
                  loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                  metrics=['accuracy'])

model.summary()
```

```python
total_epochs = INITIAL_EPOCHS + FINETUNE_EPOCHS

history_finetune = model.fit(train_dataset,
                                epochs=total_epochs,
                                initial_epoch=history.epoch[-1],
                                validation_data=val_dataset)
```

# Training Part 3
### Finetuning the Model Further by Reducing the Learning Rate by a Factor of 10

Training with a learning rate of 0.0001 resulted in a large variation in accuracy between epochs. Once the MobileNetV2 is fine tuned to a sufficient degree, we then train it at a lower learning rate to continue fine tuning and reducing the variation in performance.

```python
# Decrease learning rate
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.00001),
                  loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                  metrics=['accuracy'])

model.summary()
```

```python
total_epochs = total_epochs + FINETUNE_EPOCHS

history_finetune_lower_lr = model.fit(train_dataset,
                                          epochs=total_epochs,
                                          initial_epoch=history_finetune.epoch[-1],
                                          validation_data=val_dataset)
```

## Saving Results and Model

```python
def save_history(history, location):
    with open(location, 'w') as f:
        json.dump(str(history), f)
        
training_results = {
    'accuracy': history.history['accuracy'],
    'val_accuracy' : history.history['val_accuracy'],
    'loss' : history.history['loss'],
    'val_loss' : history.history['val_loss']
}

training_results['accuracy'] = training_results['accuracy'] + history_finetune.history['accuracy']
training_results['val_accuracy'] = training_results['val_accuracy'] + history_finetune.history['val_accuracy']
training_results['loss'] = training_results['loss'] + history_finetune.history['loss']
training_results['val_loss'] = training_results['val_loss'] + history_finetune.history['val_loss']

training_results['accuracy'] = training_results['accuracy'] + history_finetune_lower_lr.history['accuracy']
training_results['val_accuracy'] = training_results['val_accuracy'] + history_finetune_lower_lr.history['val_accuracy']
training_results['loss'] = training_results['loss'] + history_finetune_lower_lr.history['loss']
training_results['val_loss'] = training_results['val_loss'] + history_finetune_lower_lr.history['val_loss']

model.save(os.path.join('models', 'model'))
save_history(training_results, os.path.join('models', 'history'))
```

# Results

```python
def load_history(location):
    with open(location, 'r') as f:
        data_str = f.read()[1:-1]
        data_str = data_str.replace("'", '"')
        return json.loads(data_str)
    
training_results = load_history(os.path.join('models', 'history'))
loaded_model = loaded_model = tf.keras.models.load_model(os.path.join('models', 'model'))
```

```python
loss_test, accuracy_test = loaded_model.evaluate(test_dataset)
print("Accuracy on Test Dataset:", accuracy_test)
```

```python
image_batch, label_batch = test_dataset.as_numpy_iterator().next()
predictions = loaded_model.predict_on_batch(image_batch).flatten()

# Need to apply the sigmoid activation function on the predictions since it returns logits
predictions = tf.nn.sigmoid(predictions)
predictions = tf.where(predictions < 0.5, 0, 1)

print('Predictions:\n', predictions.numpy())
print('Actual:\n', label_batch)

plt.figure(figsize=(10, 10))
for ii in range(9):
    ax = plt.subplot(3, 3, ii + 1)
    plt.imshow(image_batch[ii].astype("uint8"))
    plt.title(class_names[predictions[ii]])
    plt.axis("off")
```

```python
fig, ax = plt.subplots(3,1, figsize=(10,20))

epochs_list = [ii for ii in range(1, len(training_results['accuracy'])+1)]

ax[0].plot(epochs_list, training_results['accuracy'], 'r-')
ax[0].plot(epochs_list, training_results['val_accuracy'], 'b-')
ax[0].legend(["Training Accuracy", "Validation Accuracy"])
ax[0].set_title("Training and Validation Accuracy while Training")
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("Accuracy")

ax1 = ax[0].twinx()
ax1.axvline(INITIAL_EPOCHS, ls='--')
ax1.axvline(INITIAL_EPOCHS + FINETUNE_EPOCHS, ls='--')

ax[1].plot(epochs_list, training_results['loss'], 'r-')
ax[1].plot(epochs_list, training_results['val_loss'], 'b-')
ax[1].legend(["Training Cross Entropy Loss", "Validation Cross Entropy Loss"])
ax[1].set_title("Training and Validation Cross Entropy while Training")
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("Cross Entropy")

ax1 = ax[1].twinx()
ax1.axvline(INITIAL_EPOCHS, ls='--')
ax1.axvline(INITIAL_EPOCHS + FINETUNE_EPOCHS, ls='--')

conf_matrix = confusion_matrix(label_batch, predictions.numpy())
ax[2].set_title("Confusion Matrix on the Testing Set")
ax[2].matshow(conf_matrix, cmap=plt.cm.gray)
ax[2].set_xticks(range(len(class_names)))
ax[2].set_xticklabels(class_names)
ax[2].set_yticks(range(len(class_names)))
ax[2].set_yticklabels(class_names)

f1_scores = f1_score(label_batch, predictions.numpy(), average=None)

print("COVID F1 Score:\n\t", f1_scores[0])
print("NonCOVID F1 Score:\n\t", f1_scores[1])
print("Average F1 Score:\n\t", (f1_scores[0] + f1_scores[1])/len(f1_scores))
```

## Testing the Classification of GAN Generated COVID-19 CT Scans

We selected the 19 best looking generated scans that were generated using StyleGAN2 with differential augmentation.

```python
GAN_DATASET_LOC = os.path.join("gan_dataset")
gan_dataset = tf.keras.preprocessing.image_dataset_from_directory(GAN_DATASET_LOC,
                                                                      shuffle = True,
                                                                      batch_size=BATCH_SIZE,
                                                                      image_size=IMAGE_DIM,
                                                                      seed=SEED)
```
